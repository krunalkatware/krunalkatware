```python
df = df.rename(columns={' Mean of the integrated profile': 'mean',
                        ' Standard deviation of the integrated profile': 'standard_deviation',
                        ' Excess kurtosis of the integrated profile': 'Excess_kurtoss',
                        ' Skewness of the integrated profile': 'skewness',
                        ' Mean of the DM-SNR curve': 'mean_DM-SNR_curve',
                        ' Standard deviation of the DM-SNR curve': 'std_DM-SNR_curve',
                        ' Excess kurtosis of the DM-SNR curve': 'Ex-kurtosis_DM-SNR',
                        ' Skewness of the DM-SNR curve': 'skewness_DM-SNR',
                        ' target_class': 'target_class'})
```

```python
# assuming 'y is the target variable
X = df.drop(columns=['target_class'])
y = df['target_class']
```

```python
#One-hot encode categorical variables
#X = pd.get_dummies(X)
X = pd.get_dummies(X, dtype=int)
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, train_size=0.7,
                                               test_size=0.3, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
#import metrics to comput accuracy
from sklearn.metrics import accuracy_score
svc = SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Model accuracy score with default hyperparameters:{0:0.4f}'
.format(accuracy_score(y_test,y_pred)))
```

```python
svc = SVC(C=100.0)
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Model accuracy score with rbf kernal and c=100.0 hyperparameters:{0:0.4f}'
.format(accuracy_score(y_test,y_pred)))
```

```python
svc = SVC(C=1000.0)
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Model accuracy score with rbf kernal and c=1000.0 hyperparameters:{0:0.4f}'
.format(accuracy_score(y_test,y_pred)))
```

```python
svc = SVC(kernel='linear', C=1.0)
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Model accuracy score with linear kernal and c=1.0 hyperparameters:{0:0.4f}'
.format(accuracy_score(y_test,y_pred)))
```
